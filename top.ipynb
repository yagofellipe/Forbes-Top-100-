```python
# Importando bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("dark")
```

```python
# Importando dataset
data = pd.read_csv('forbes_2022_billionaires.csv', sep = ",")
df = pd.DataFrame(data)
df
```

```python
# removendo uma coluna
df.drop('country', axis = 1, inplace = True) 
```

```python
# verificando a quantidade de valores nulos

df.isnull().sum()
```

```python
# MÃ©dia mundial de acordo com a idade
fortuna_mundial_media = df.groupby('age')['finalWorth'].mean()
fortuna_mundial_media 
```

```python
# MÃ©dia brasileira de acordo com idade
df_brasil = df[df['countryOfCitizenship']=='Brazil'].copy()
fortuna_brasil_media = df_brasil.groupby('age')['finalWorth'].mean()
fortuna_brasil_media
```

```python
# MÃ©dia mundial e mÃ©dia brasileira
df['finalWorth'].mean
df_brasil['finalWorth'].mean()
```

```python
# quantidade de bilionarios por setor
df['category'].value_counts()
```

```python
# quantidade de bilionarios por setor no Brasil

####### fazer um grÃ¡fico de pizza com porcetentagem e discursar como o setor de tech nÃ£o Ã© desenvolvido
df_brasil['category'].value_counts()
```

```python
anos = df_brasil.age.unique()
fortuna_brasil_media = df_brasil.groupby('age')['finalWorth'].mean()
fortuna_mundial_media = df.groupby('age')['finalWorth'].mean()


fig = plt.figure(figsize=(15,5))
ax = sns.lineplot(x=anos,y=fortuna_mundial_media, label='Mundial', color='blue')
ax = sns.lineplot(x=anos, y = fortuna_brasil_media, label='Brasil', color='green')
plt.title('MÃ©dia de suicÃ­dio ao longo do tempo (Brasil X Mundo)', fontsize=19)
plt.ylabel('NÂ° de casos a cada 100 mil pessoas',fontsize=13)
```
